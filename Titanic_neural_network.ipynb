```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
train_data = pd.read_csv('/content/train.csv')
train_data
```

```python
test_data = pd.read_csv('/content/test.csv')
test_data
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
train_data = train_data.drop(['Name',  'Cabin'], axis=1)
test_data = test_data.drop(['Name',  'Cabin'], axis=1)
```

```python
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
train_data['Fare'].fillna(train_data['Fare'].mean(), inplace=True)
```

```python
test_data['Age'].fillna(test_data['Age'].mean(), inplace=True)
test_data['Embarked'].fillna(test_data['Embarked'].mode()[0], inplace=True)
test_data['Fare'].fillna(test_data['Fare'].mean(), inplace=True)
```

```python
train_data
```

```python
test_data
```

```python
test_data["Sex"],class_name = pd.factorize(test_data["Sex"])
test_data["Ticket"],class_name = pd.factorize(test_data["Ticket"])
test_data["Embarked"],class_name = pd.factorize(test_data["Embarked"])
```

```python
train_data["Sex"],class_name = pd.factorize(train_data["Sex"])
train_data["Ticket"],class_name = pd.factorize(train_data["Ticket"])
train_data["Embarked"],class_name = pd.factorize(train_data["Embarked"])
```

```python
test_data
```

```python
train_data
```

```python
X_train = train_data.drop("Survived", axis=1)
y_train = train_data["Survived"]
```

```python
#X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=34, validation_split=0.2)
```

```python
predictions = model.predict(test_data)
predictions = (predictions > 0.5).astype(int)
```

```python
submission_df = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": predictions.flatten()
})
```

```python
submission_df.to_csv("submission.csv", index=False)
```
